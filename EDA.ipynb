```python
from sklearn.model_selection import train_test_split
import lightgbm as lgb
```

```python
X = df.drop('Attrition',axis=1)
y = df['Attrition'] # 目的変数
# トレーニングデータ,テストデータの分割
X_train, X_valid, y_train, y_valid = train_test_split(X, y,test_size=0.2, random_state=0)
```

```python
import optuna
import lightgbm as lgb
from sklearn.metrics import recall_score, roc_auc_score
import numpy as np

def objective(trial):
    params = {
        'objective': 'binary',
        'metric': 'binary_logloss',
        'boosting_type': 'gbdt',
        'num_leaves': trial.suggest_int('num_leaves', 10, 100),
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-3, 1.0),
        'feature_fraction': trial.suggest_uniform('feature_fraction', 0.1, 1.0),
        'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.1, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 1, 10),
        'min_child_samples': trial.suggest_int('min_child_samples', 1, 100),
        'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
        'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),
        'scale_pos_weight': trial.suggest_uniform('scale_pos_weight', 1, 10),
    }

    train_data = lgb.Dataset(X_train, label=y_train)
    valid_data = lgb.Dataset(X_valid, label=y_valid, reference=train_data)

    callbacks = [lgb.early_stopping(stopping_rounds=50)]

    model = lgb.train(params, train_data, num_boost_round=1000,
                      valid_sets=[valid_data], callbacks=callbacks)

    preds = model.predict(X_valid)
    recall = recall_score(y_valid, preds > 0.5)
    auc = roc_auc_score(y_valid, preds)

    return recall * 0.7 + auc * 0.3
# Optunaの実行
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# 最適なパラメータの取得
best_params = study.best_params
print("Best parameters:", best_params)

# 最適なパラメータでモデルを再学習
best_model = lgb.train(best_params, lgb.Dataset(X_train, label=y_train), num_boost_round=1000,
                       valid_sets=[lgb.Dataset(X_valid, label=y_valid)],
                       callbacks=[lgb.early_stopping(stopping_rounds=50)])

# 検証データでの予測
y_pred = best_model.predict(X_valid)

# しきい値の調整（前回のコードと同様）
thresholds = np.arange(0.1, 0.5, 0.01)
recalls = [recall_score(y_valid, y_pred >= threshold) for threshold in thresholds]
target_recall = 0.9
best_threshold = thresholds[np.argmin(np.abs(np.array(recalls) - target_recall))]

print(f"新しいしきい値: {best_threshold:.2f}")
y_pred_binary = (y_pred >= best_threshold).astype(int)

# 結果の評価
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_valid, y_pred_binary))
print(confusion_matrix(y_valid, y_pred_binary))

# 特徴量の重要度を表示
importance = best_model.feature_importance()
feature_names = best_model.feature_name()

for name, importance in sorted(zip(feature_names, importance), key=lambda x: x[1], reverse=True):
    print(f"{name}: {importance}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# データの読み込み（dfは既に読み込まれていると仮定）

# インセンティブの最大値を取得し、ビンの数を決定
max_incentive = df['Incentive'].max()
num_bins = 10

# 0から最大値までを等間隔で分割
bins = np.linspace(0, max_incentive, num_bins + 1)

# インセンティブを0からの適切な範囲でビン分割
df['Incentive_Bin'] = pd.cut(df['Incentive'], bins=bins, include_lowest=True)

# インセンティブ範囲ごとの離職率を計算
attrition_rate = df.groupby('Incentive_Bin')['Attrition'].mean().reset_index()
attrition_rate['Attrition_Rate'] = attrition_rate['Attrition'] * 100  # パーセンテージに変換

# グラフの設定
plt.figure(figsize=(14, 8))

# 棒グラフの作成
sns.barplot(x='Incentive_Bin', y='Attrition_Rate', data=attrition_rate)

# グラフのタイトルと軸ラベルの設定
plt.title("インセンティブ範囲ごとの離職率", fontsize=16)
plt.xlabel("インセンティブ範囲", fontsize=12)
plt.ylabel("離職率 (%)", fontsize=12)

# x軸のラベルを回転
plt.xticks(rotation=45, ha='right')

# y軸の範囲を0-100%に設定
plt.ylim(0, 100)

# 棒グラフの上に離職率の値を表示
for i, v in enumerate(attrition_rate['Attrition_Rate']):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom')

# グラフの余白を調整
plt.tight_layout()

# グラフの表示
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# データの読み込み（dfは既に読み込まれていると仮定）

# Attritionカラムの値を日本語のラベルにマッピング
attrition_labels = {0: "離職していない", 1: "離職した"}
df["Attrition_Label"] = df["Attrition"].map(attrition_labels)

# グループごとの年齢の平均を計算
age_means = df.groupby("Attrition_Label")["Age"].mean().reset_index()

# データフレームの順序を逆にする
age_means = age_means.iloc[::-1]

# グラフの設定
plt.figure(figsize=(10, 6))

# 棒グラフの作成（順序を指定）
sns.barplot(x="Attrition_Label", y="Age", data=age_means, order=["離職していない", "離職した"], palette=['#1f77b4', '#ff7f0e'])

# グラフのタイトルと軸ラベルの設定
plt.title("離職状況別の平均年齢", fontsize=16)
plt.xlabel("離職状況", fontsize=14)
plt.ylabel("平均年齢", fontsize=14)

# 棒の上に平均年齢の値を表示
for i, v in enumerate(age_means["Age"]):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=12)

# Y軸の範囲を調整（0から始まるようにする）
plt.ylim(0, max(age_means["Age"]) * 1.1)

# グラフの余白を調整
plt.tight_layout()

# グラフの表示
plt.show()

# 平均年齢の差を計算して表示
age_diff = age_means.loc[age_means["Attrition_Label"] == "離職した", "Age"].values[0] - \
           age_means.loc[age_means["Attrition_Label"] == "離職していない", "Age"].values[0]
print(f"離職したグループと離職していないグループの平均年齢の差: {age_diff:.2f} 歳")
```
